```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.font_manager

from matplotlib import rc

plt.style.use(['science', 'ieee', 'grid'])

tex_fonts = {
    'text.usetex': True,
    'font.family': 'serif',
    'font.serif': 'Times',
    'axes.labelsize': 10,
    'font.size': 10,
    'legend.fontsize': 8,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8
}
plt.rcParams.update(tex_fonts)
```

```python
import sys
def set_size(width, fraction=1, subplots=(1, 1), ratio = (5**.5 - 1) / 2):
    """Set figure dimensions to avoid scaling in LaTeX.

    Parameters
    ----------
    width: float or string
            Document width in points, or string of predined document type
    fraction: float, optional
            Fraction of the width which you wish the figure to occupy
    subplots: array-like, optional
            The number of rows and columns of subplots.
    Returns
    -------
    fig_dim: tuple
            Dimensions of figure in inches
    """
    if width == 'usenix_textwidth':
        width_pt = 505.89
    elif width == 'usenix_linewidth':
        width_pt = 241.02039
    elif width == 'acm_columnwidth':
        width_pt = 395.8225
    else:
        print('unrecognized width')
        sys.exit(0)

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    # https://disq.us/p/2940ij3
    golden_ratio = ratio

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

```python
def autolabel(rects, itype = 'float', fontsize = 10):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        if itype == 'float':
            ax.annotate('%.1f' % height,
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize = fontsize)
        elif itype == 'int':
            ax.annotate('%d' % height,
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize = fontsize)
```

```python
df = pd.read_csv('./recovery_perf.csv')

parallelism = [1, 2, 4]
perf = [df.loc[(df['PARALLELISM'] == p) & (df['OSIRIS_LIMIT'] == 4)]['TOTAL\n(s)'].mean() for p in parallelism]

# impact of parallelism on recovery time
fig, ax = plt.subplots(1, 1, figsize = set_size('acm_columnwidth', ratio = .3, fraction = .49))
rects1 = ax.bar(np.arange(len(parallelism)), perf, width = 0.25)
ax.set_xticks(np.arange(len(parallelism)))
ax.set_xticklabels(parallelism)
# ax.set_xlabel('Parallelism')
ax.set_ylabel('Time (ms)')
# ax.set_title('Impact of Parallelism on Recovery Time')
ax.plot(np.arange(len(parallelism)), perf, 'x-', color = 'r')

ax.set_ylim(0, 2000)
autolabel(rects1)

fig.savefig('/home/yzou/Dropbox/Research-Yu-Zou-2020/2020-ATC-MemSec-Zou-v2/Zou/plots/impact_parallelism_on_recovery.pdf', dpi=300, bbox_inches='tight', format='pdf')

# impact of osiris_limit on recovery time
osiris_limit = [2, 4, 8, 16]
perf = [df.loc[(df['OSIRIS_LIMIT'] == o) & (df['PARALLELISM'] == 2)]['TOTAL\n(s)'].mean() for o in osiris_limit]

fig, ax = plt.subplots(1, 1, figsize = set_size('acm_columnwidth', ratio = .3, fraction = .49))
rects1 = ax.bar(np.arange(len(osiris_limit)), perf, width = 0.35)
ax.set_xticks(np.arange(len(osiris_limit)))
ax.set_xticklabels(osiris_limit)
# ax.set_xlabel('Osiris Limit')
ax.set_ylabel('Time (ms)')
# ax.set_title('Impact of Osiris Limit on Recovery Time')

ax.set_ylim(0, 3000)
autolabel(rects1)

fig.savefig('/home/yzou/Dropbox/Research-Yu-Zou-2020/2020-ATC-MemSec-Zou-v2/Zou/plots/impact_osiris_limit_on_recovery.pdf', dpi=300, bbox_inches='tight', format='pdf')
```

```python
df = pd.read_csv('./impact_osiris_limit_64B.csv')

# impact of osiris_limit on runtime performance
osiris_limit = [2, 4, 8, 16]
seq_write_time = np.array([int(df.loc[(df['APP'] == 'seq-write') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['MIN'], base=16)
             for o in osiris_limit])
seq_read_time = np.array([int(df.loc[(df['APP'] == 'seq-read') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['MIN'], base=16)
             for o in osiris_limit])
rand_write_time = np.array([int(df.loc[(df['APP'] == 'rand-write') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['MIN'], base=16)
             for o in osiris_limit])
rand_read_time = np.array([int(df.loc[(df['APP'] == 'rand-read') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['MIN'], base=16)
             for o in osiris_limit])

fig, ax = plt.subplots(1, 1, figsize = set_size('acm_columnwidth', ratio = .4, fraction = .49))
width = 0.35
rects1 = ax.bar(np.arange(len(osiris_limit))-.5*width, seq_write_time/1e8, width)
rects2 = ax.bar(np.arange(len(osiris_limit))+.5*width, rand_write_time/1e8, width)
ax.set_xticks(np.arange(len(osiris_limit)))
ax.set_xticklabels(osiris_limit)
# ax.set_xlabel('Osiris Limit')
ax.set_ylabel('Time (s)')
# ax.set_title('Impact of Osiris Limit on Runtime Performance')
ax.legend(['Sequential Write', 'Random Write'])

autolabel(rects1, fontsize = 8)
autolabel(rects2, fontsize = 8)
ax.set_ylim(0, 70)
# fig.tight_layout()
fig.savefig('/home/yzou/Dropbox/Research-Yu-Zou-2020/2020-ATC-MemSec-Zou-v2/Zou/plots/osiris_limit_runtime.pdf', dpi=300, bbox_inches='tight', format='pdf')

# impact of osiris_limit on number of counter persistence
osiris_limit = [2, 4, 8, 16]
seq_write_npersists = np.array([df.loc[(df['APP'] == 'seq-write') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['NCOUNTER_PERSIST']
             for o in osiris_limit])
seq_read_npersists = np.array([df.loc[(df['APP'] == 'seq-read') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['NCOUNTER_PERSIST']
             for o in osiris_limit])
rand_write_npersists = np.array([df.loc[(df['APP'] == 'rand-write') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['NCOUNTER_PERSIST']
             for o in osiris_limit])
rand_read_npersists = np.array([df.loc[(df['APP'] == 'rand-read') & (df['OSIRIS_LIMIT'] == o) & (df['TYPE'] == '64B')].iloc[0]['NCOUNTER_PERSIST']
             for o in osiris_limit])

fig, ax = plt.subplots(1, 1, figsize = set_size('acm_columnwidth', ratio = .4, fraction=.49))
width = 0.35
rects1 = ax.bar(np.arange(len(osiris_limit))-.5*width, seq_write_npersists, width)
rects2 = ax.bar(np.arange(len(osiris_limit))+.5*width, rand_write_npersists, width)
ax.set_xticks(np.arange(len(osiris_limit)))
ax.set_xticklabels(osiris_limit)
# ax.set_xlabel('Osiris Limit')
ax.set_ylabel('Counter Writes')
# ax.set_title('Impact of Osiris Limit on Number of Counter Persistence')
ax.legend(['Sequential Write', 'Random Write'])

autolabel(rects1, 'int', 8)
autolabel(rects2, 'int', 8)
ax.set_ylim(0, 1.4*1e6)
# fig.tight_layout()
fig.savefig('/home/yzou/Dropbox/Research-Yu-Zou-2020/2020-ATC-MemSec-Zou-v2/Zou/plots/osiris_limit_lifetime.pdf', dpi=300, bbox_inches='tight', format='pdf')
```
